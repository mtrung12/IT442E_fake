```python
import pandas as pd
import json


df = pd.read_csv('saved_data/df_clean.csv')

hierarchy = {}

for index, row in df.iterrows():
    city = str(row['city'])
    district = str(row['district'])
    ward = str(row['ward'])
    street = str(row['street'])
    if city not in hierarchy:
        hierarchy[city] = {}
    if district not in hierarchy[city]:
        hierarchy[city][district] = {}
    if ward not in hierarchy[city][district]:
        hierarchy[city][district][ward] = []
    if street not in hierarchy[city][district][ward]:
        hierarchy[city][district][ward].append(street)


with open('saved_data/vn_available_locations.json', 'w', encoding='utf-8') as f:
    json.dump(hierarchy, f, ensure_ascii=False, indent=4)
```
